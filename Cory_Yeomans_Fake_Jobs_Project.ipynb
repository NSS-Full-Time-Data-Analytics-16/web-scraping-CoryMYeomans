```python
import requests
```

```python
from bs4 import BeautifulSoup as BS
```

```python
import pandas as pd
```

```python
URL = 'https://realpython.github.io/fake-jobs/'
```

```python
response = requests.get(URL)
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.find('h2').text
```

```python
job_title_tags = soup.find_all("h2")

job_titles = [tag.text for tag in job_title_tags]

print(job_titles)
```

```python
job_company_tags = soup.find_all("h3")

job_company = [tag.text for tag in job_company_tags]

print(job_company)
```

```python
job_location_tags = soup.find_all("p", class_='location')

job_location = [tag.text.strip() for tag in job_location_tags]

print(job_location)
```

```python
job_date_tags = soup.find_all("time")

job_date = [tag.text.strip() for tag in job_date_tags]

print(job_date)
```

```python
jobs_df = pd.DataFrame({"Title": job_titles, "Company": job_company, "Location": job_location, "Date Posted": job_date})
jobs_df
```

```python
job_url_tags = soup.find_all("a", class_='card-footer-item')

job_url = [tag['href'] for tag in job_url_tags if tag.text.strip() == 'Apply']

print(job_url)
```

```python
jobs_df['url'] = job_url
jobs_df
```

```python
response2 = requests.get(jobs_df['url'][0])
```

```python
soup2 = BS(response2.text)
```

```python
print(soup2.prettify())
```

```python
job_description_tags = soup2.find("div", class_='content').p.text.strip()
print(job_description_tags)
```

```python
def get_desc(url):
    response2 = requests.get(url)
    soup2 = BS(response2.text)
    job_description_tags = soup2.find_all("div", class_='content')
    job_description = job_description_tags[0].p.text.strip() #[tag.p.text.strip() for tag in job_description_tags]
    return job_description
```

```python
def get_desc(url):
    response2 = requests.get(url)
    soup2 = BS(response2.text)
    job_description = soup2.find("div", class_='content').p.text.strip()
    return job_description
```
